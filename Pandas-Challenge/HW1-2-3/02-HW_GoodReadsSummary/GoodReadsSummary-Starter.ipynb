```python
# Import Dependencies
import pandas as pd
import os
```

```python
# Path and filename for the file to Load
fcsv_pth = 'Resources/books_clean.csv'


# Read the modified GoodReads csv and store into Pandas DataFrame

f_df = pd.read_csv(fcsv_pth)
f_df.head()
```

```python
# Calculate the number of unique authors in the DataFrame
authcnt = f_df['Authors'].nunique()
print(f'There are {authcnt} unique authors in this list.')

# Calculate the earliest/latest year a book was published
earlpub = f_df['Publication Year'].min()
latpub = f_df['Publication Year'].max()

print(f'The earliest book was published in {earlpub}.\nThe latest book was published in {latpub}')

# Calculate the total reviews column that contains the sum of reviews (total - all authors)
ttl_rev_srs = f_df[['One Star Reviews','Two Star Reviews','Three Star Reviews','Four Star Reviews','Five Star Reviews']].sum(axis=1)
f_df['Total Reviews'] = ttl_rev_srs
f_df.head()
# Hint: use the pandas' sum(axis='columns') to get the sum across the columns 
```

```python
#Create a total_reviews variable that sums up the values from your new Total Reviews column in the DataFrame
ttl_revw = f_df['Total Reviews'].sum()
print(f'The total reviews for all authors is {ttl_revw}')
```

```python
# Place all of the data found into a summary DataFrame
summary_df = pd.DataFrame(columns =['Total Unique Authors','Earliest Year','Latest Year','Total Reviews']) 
summary_df['Total Unique Authors'] = [authcnt]
summary_df['Earliest Year'] = [earlpub]
summary_df['Latest Year'] = [latpub]
summary_df['Total Reviews'] = [ttl_revw]

# Show the DataFrame
summary_df
```
