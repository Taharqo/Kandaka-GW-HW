# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school. 
---

```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to Load 
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset (consider using a left join)
schl_std_data_df = pd.merge(school_data,student_data,on = 'school_name')
```

## District Summary

```python
# Calculate the Totals (Schools and Students)
ttl_std = len(student_data)
ttl_schl = len(school_data)

# Calculate the Total Budget
ttl_bdgt = school_data['budget'].sum()

# Calculate the Average Scores
avg_rding_scr = student_data['reading_score'].mean()
avg_mth_scr = student_data['math_score'].mean()

# Calculate the Percentage Pass Rates
stds_passing_mth_s = student_data['math_score'] >= 70
stds_passing_mth = student_data.loc[stds_passing_mth_s]
stds_prc_pass_mth = (len(stds_passing_mth)/ttl_std)*100

stds_passing_rding_s = student_data['reading_score'] >= 70
stds_passing_rding = student_data.loc[stds_passing_rding_s]
stds_prc_pass_rding = (len(stds_passing_rding)/ttl_std)*100

stds_passing_mth_rding_s = (student_data['reading_score'] >= 70) & (student_data['math_score'] >= 70)
stds_passing_mth_rding = student_data.loc[stds_passing_mth_rding_s]
stds_prc_pass_mth_rding= (len(stds_passing_mth_rding)/ttl_std) * 100

# Minor Data Cleanup


```

```python
# Display the data frame
data = {
        'Total Schools':[ttl_schl],
        'Total Studens':[ttl_std],
        'Total Budget':[ttl_bdgt],
        'Average Math Score':[avg_mth_scr],
        'Average Reading Score':[avg_rding_scr],
        '% Passing Math':[stds_prc_pass_mth],
        '% Passing Reading':[stds_prc_pass_rding],
        '% Overall Passing Rate':[stds_prc_pass_mth_rding]
       }

ovrall_summary_df = pd.DataFrame(data)

ovrall_summary_df.style.format({
                                'Total Studens':'{:,}', 
                                'Total Budget':'${:,.02f}',
                                'Average Math Score':'{:,.06f}',
                                'Average Reading Score':'{:,.05f}',
                                '% Passing Math': '{:,.06f}',
                                '% Passing Reading':'{:,.06f}',
                                '% Overall Passing Rate':'{:,.06f}'
                              })
```

## School Summary

```python
# Calculate the total student count
grp_by_schl = schl_std_data_df.groupby(['school_name'])
stds_by_schl = grp_by_schl['student_name'].count()
stds_by_schl
```

```python
# Determine the School Type
#schol_type = school_data[['school_name','type']].set_index('school_name')
schl_type = grp_by_schl['type'].unique()
schl_type = schl_type.to_string()
schol_type
```

```python
# Calculate the total school budget and per capita spending

budget_by_schl = grp_by_schl['budget'].unique()
budget_by_schl = budget_by_schl.astype('float64')
ttl_stds_by_schl = grp_by_schl['student_name'].count()
bdg_per_std = budget_by_schl/ttl_stds_by_schl
bdg_per_std=bdg_per_std.astype('float64')
```

```python
# Calculate the average test scores
avg_reading_scr_by_schl = grp_by_schl['reading_score'].mean()
avg_mth_scr_by_schl = grp_by_schl['math_score'].mean()
```

```python
# Calculate the passing scores by creating a filtered data frame
passing_rding_filtr = schl_std_data_df['reading_score']>= 70 
passing_rding_stds = schl_std_data_df.loc[passing_rding_filtr]
passing_rding_by_schol_obj = passing_rding_stds.groupby(['school_name'])
passing_rding_by_schol = passing_rding_by_schol_obj['reading_score'].count()
prc_passing_rding = (passing_rding_by_schol/stds_by_schl)*100

passing_mth_filter_s = schl_std_data_df['math_score']>=70
passing_mth_stds_df = schl_std_data_df.loc[passing_mth_filter_s]
passing_mth_by_schol_obj = passing_mth_stds_df.groupby(['school_name'])
passing_mth_by_scol = passing_mth_by_schol_obj['math_score'].count()
prc_passing_mth = (passing_mth_by_scol/stds_by_schl)*100

# stds_passed_mth_rding_both = passing_mth_stds_df+passing_rding_stds
# print(len(stds_passed_mth_rding_both))

stds_passing_mth_rding = (schl_std_data_df['reading_score']>=70) & (schl_std_data_df['math_score']>=70)
pssing_mth_rding_stds = schl_std_data_df.loc[stds_passing_mth_rding]
passing_mth_rding_by_schl_obj = pssing_mth_rding_stds.groupby(['school_name'])
passing_mth_rding_by_schl = passing_mth_rding_by_schl_obj['reading_score'].count()
prc_passing_rding_mth = (passing_mth_rding_by_schl/stds_by_schl)*100

#Convert to data frame
values_dict = {
                'School Type':schl_type,
                'Total Students':stds_by_schl,
                'Total Budget':budget_by_schl, 
                'Budget per Student':bdg_per_std, 
                'Average Reading Score':avg_reading_scr_by_schl,
                'Average Math Score':avg_mth_scr_by_schl,
                '% Passing Reading Score':prc_passing_rding,
                '% Passing Math Score':prc_passing_mth,
                '% Passing Reading & Math Score':prc_passing_rding_mth  
                }

summary_by_school = pd.DataFrame(values_dict)

summary_by_school.style.format({
                                'Total Studens':'{:,}', 
                                'Total Budget':'${:,.02f}',
                                'Budget per Student':'${:,.02f}',
                                'Average Math Score':'{:,.06f}',
                                'Average Reading Score':'{:,.06f}',
                                '% Passing Math': '{:,.06f}%',
                                '% Passing Reading':'{:,.06f}%',
                                '% Overall Passing Rate':'{:,.06f}%'
                              })
# Minor data munging

# Display the data frame
```
