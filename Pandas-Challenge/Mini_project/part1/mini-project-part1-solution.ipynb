```python
import pandas as pd
```

```python
fpath_csv = 'Resources/2016-FCC-New-Coders-Survey-Data.csv'
```

```python
df = pd.read_csv(fpath_csv)
df.head()
```

```python
#Return a subset of the DataFrame
subset_col = [0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]
subset_df = df.iloc[:,subset_col]
subset_df.head()
```

```python
#replace 0.0 with no & 1.0 with yes across the entire DataFrame
switch = {0.0:'no',1.0:'yes'}
mod_subset_df = subset_df.replace(switch)
mod_subset_df.head()
```

```python
#Calculate the total number of respondents in the sub-table you built.
ttl_resp = len(mod_subset_df)
print(f'Total Respondents: {ttl_resp}')
```

```python
#Create a table out of the rows corresponding only to people who did attend a bootcamp.
att_yes_s = mod_subset_df['AttendedBootcamp'] == 'yes'
att_yes_df = mod_subset_df[att_yes_s]
att_yes_df.head()
```

```python
#Calculate the number of people who attended a bootcamp.
ttl_bc_yes = len(att_yes_df)
print(f'Total bootcamp attendees: {ttl_bc_yes}')
```

```python
#Calculate the average age of bootcamp attendees.
avg_age = att_yes_df['Age'].mean()
print(f'Attendee average age: {avg_age} years')
```

```python
#Calculate the number of bootcamp attendees who self-identify as male; female; or neither.
yes_m_s = att_yes_df['Gender'] == 'male'
ttl_m_s = att_yes_df['Gender'].loc[yes_m_s]
count_m = ttl_m_s.count()
print(f'Total males attended: {count_m}')

yes_f_s = att_yes_df['Gender'] == 'female'
ttl_f_s = att_yes_df['Gender'].loc[yes_f_s]
count_f = ttl_f_s.count()
print(f'Total females attended: {count_f}')

yes_nb = (att_yes_df['Gender'] != 'male')&(att_yes_df['Gender'] != 'female')
ttl_nb_s = att_yes_df['Gender'].loc[yes_nb]
count_nb = ttl_nb_s.count()
print(f'Total non-binary attended: {count_nb}')
```

```python
#Calculate the number of bootcamp attendees who hold college degrees.
#att_yes_df['SchoolDegree'].unique()

degree =  (att_yes_df['SchoolDegree']== "bachelor's degree")|\
          (att_yes_df['SchoolDegree']== "master's degree (non-professional)")|\
          (att_yes_df['SchoolDegree']== 'professional degree (MBA, MD, JD, etc.)')|\
          (att_yes_df['SchoolDegree']== "associate's degree")|\
          (att_yes_df['SchoolDegree']==  'Ph.D.')
loc_deg_df = att_yes_df.loc[degree]
deg_cnt = len(loc_deg_df)
print(deg_cnt)
```

```python
#Calculate the percentage of respondents who attended a bootcamp.
prct_att_yes = (ttl_bc_yes/ttl_resp)*100
print(f'{prct_att_yes}% of respondants attended a bootcamp')
```

```python
#Calculate the percentage of people who attended a bootcamp and hold a college degree.
prct_w_deg = (deg_cnt/ttl_bc_yes)*100
print(f'{prct_w_deg}% of Attendees have degrees.')
```

```python
#Calculate the average post-bootcamp salary.
pstbc_salary = att_yes_df['BootcampPostSalary'].mean()
pstbc_salary 
```

```python
#Create a new, two-row table collecting the above data.
data = {'Total Surveyed':[ttl_resp],
        'Total Bootcamp attendees':[ttl_bc_yes],
        '% attended bootcamp':[prct_att_yes],
        'Avg Age':[avg_age],
        'Has Degree':[prct_w_deg],
        '%Male':[(count_m/ttl_bc_yes)*100],
        '%Female':[(count_f/ttl_bc_yes)*100],
        '%None Gender Specific':[(count_nb/ttl_bc_yes)*100],
        'Average Post Bootcamp Salary':[pstbc_salary ]}
summary_df = pd.DataFrame(data)
summary_df
```

```python
#Use the format method to prettify your tableâ€”
#i.e., use format to display percents as percents; 
#display numbers to a reasonable number of decimal points; etc.
summary_df = summary_df.style.format({
                            '% attended bootcamp':'{:,.02f}%',
                            'Avg Age':'{:,.02f}',
                            'Has Degree':'{:,.02f}',
                            '%Male':'{:,.02f}%',
                            '%Female':'{:,.02f}%',
                            '%None Gender Specific':'{:,.02f}%',
                            'Average Post Bootcamp Salary':'{:,.02f}'
                        })
```

```python
#Finally, export the final table into an Excel file.
summary_df.to_csv('bootcamp_summar.csv',index=True)
```
