```python
import pandas as pd
```

```python
fpath_csv = 'Resources/2016-FCC-New-Coders-Survey-Data.csv'
df = pd.read_csv(fpath_csv)
df.head()
```

```python
# Create a new table using the following columns: 
#[0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111].
columns = [0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]
subset_df = df.iloc[:,columns]
subset_df.head() 
```

```python
# The dataset currently uses 0.0 to represent "No" or "False", and "1.0" to represent "Yes"/"True". 
#Replace all instances of "0.0" with No, and all instances of "1.0" with "Yes".
switch = {0.0:'no',1.0:'yes'}
subset_switch_df = subset_df.replace(switch)
subset_switch_df.head()
```

```python
# Extract rows corresponding only to respondents who attended a bootcamp.
att_yes_s = subset_switch_df['AttendedBootcamp'] == 'yes'
att_yes_df = subset_switch_df.loc[att_yes_s]
att_yes_df.head()
```

```python
# Create a DataFrame with two columns: One with the bootcamp name, 
#and one with the number of respondents who went to each bootcamp.
bc_num_att_df = att_yes_df.groupby(['BootcampName'])['AttendedBootcamp'].count().reset_index()
bc_num_att_df.head()
```

```python
# Create another DataFrame with two columns: One for the bootcamp name, 
#and one for the number of respondents who recommend it.
swap = {'no':0,'yes':1}
swap_recom_df = att_yes_df.replace(swap)
select = swap_recom_df['BootcampRecommend']== 1
swap_recom_df.loc[select]
swap_recom_yes_df = swap_recom_df.groupby(['BootcampName'])['BootcampRecommend'].sum().reset_index()
swap_recom_yes_df.head()
```

```python
# Create a new DataFrame by merging the previous two DataFrames on the appropriate column.
merged_df = pd.merge(swap_recom_yes_df,bc_num_att_df, on = 'BootcampName')
merged_df.head()
```

```python
# In your new DataFrame, create a new column containing the percentage of respondents 
#for each bootcamp who would recommend that bootcamp.
prct_recom = (merged_df['BootcampRecommend']/merged_df['AttendedBootcamp'])*100
merged_df['% of Recomendors'] = prct_recom
merged_df.head()
```

```python
# Sort the new DataFrame in descending order of the percentage of recommenders you just calculated.
Summar_sort_df = merged_df.sort_values(by = ['% of Recomendors'],ascending = False)
Summar_sort_df.head()
```
